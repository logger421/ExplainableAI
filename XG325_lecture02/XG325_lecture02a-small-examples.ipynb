### Anscombe's quartet

```python
import numpy
import pandas
import sklearn.linear_model
import sklearn.metrics

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = numpy.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]).reshape(-1, 1)
y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
x4 = [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8]

x4 = numpy.array([8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8]).reshape(-1, 1)
y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]
```

```python
# Generate Dataset
reg1 = sklearn.linear_model.LinearRegression()
reg1.fit(x, y1)
reg2 = sklearn.linear_model.LinearRegression()
reg2.fit(x, y2)

# Plot
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(x, y1, 'bo'); axes[0].plot(x, reg1.predict(x), 'b-');
axes[1].plot(x, y2, 'go'); axes[1].plot(x, reg2.predict(x), 'g-'); plt.show()
```

```python
# Generate Dataset
reg3 = sklearn.linear_model.LinearRegression()
reg3.fit(x, y3)
reg4 = sklearn.linear_model.LinearRegression()
reg4.fit(x4, y4)

# Plot
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(x, y3, 'ro'); axes[0].plot(x, reg3.predict(x), 'r-');
axes[1].plot(x4, y4, 'ko'); axes[1].plot(x4, reg4.predict(x4), 'k-'); plt.show()
```

```python
reg1.score(x, y1), reg2.score(x, y2), reg3.score(x, y3), reg4.score(x4, y4)
```

```python
msq = sklearn.metrics.mean_squared_error
msq(reg1.predict(x), y1), msq(reg2.predict(x), y2), msq(reg3.predict(x), y3), msq(reg4.predict(x4), y4)
```

```python
mae = sklearn.metrics.mean_absolute_error
mae(reg1.predict(x), y1), mae(reg2.predict(x), y2), mae(reg3.predict(x), y3), mae(reg4.predict(x4), y4)
```

```python
me = sklearn.metrics.max_error
me(reg1.predict(x), y1), me(reg2.predict(x), y2), me(reg3.predict(x), y3), me(reg4.predict(x), y4)
```
