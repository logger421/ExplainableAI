### Generate Data

```python
import numpy
import pandas
import sklearn.linear_model
import sklearn.metrics

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load dataset
x = numpy.linspace(start=1, stop=3, num=201).reshape(-1, 1)
y = 0.56*x**3 - 4*x**2 + 10*x - 5
y += numpy.random.normal(0, 0.1, size=(201, 1))
```

```python
# Plot
plt.plot(x, y, 'ko')
plt.xlabel('number of units (in thousands)'); plt.ylabel('cost (in thousands $)'); plt.show()
```

### Linear Regression

```python
# Fit
reg1 = sklearn.linear_model.LinearRegression()
reg1.fit(x, y)

# Plot results
plt.plot(x, y, 'ko')
plt.plot(x, reg1.predict(x), 'ro')
plt.xlabel('number of units (in thousands)'); plt.ylabel('cost (in thousands $)'); plt.show()
```

```python
reg1.intercept_, reg1.coef_
```

```python
x2 = pandas.DataFrame.from_dict({'x': x.flatten(), 'x2': (x*x).flatten()})
x2
```

```python
# Fit
reg2 = sklearn.linear_model.LinearRegression()
reg2.fit(x2, y)

# Plot results
plt.plot(x, y, 'ko')
plt.plot(x, reg2.predict(x2), 'ro')
plt.xlabel('number of units (in thousands)'); plt.ylabel('cost (in thousands $)'); plt.show()
```

```python
reg2.intercept_, reg2.coef_
```

```python
x3 = pandas.DataFrame.from_dict({
    'x': x.flatten(),
    'x2': (x*x).flatten(),
    'x3': (x*x*x).flatten()})
x3
```

```python
# Fit
reg3 = sklearn.linear_model.LinearRegression()
reg3.fit(x3, y)

# Plot results
plt.plot(x, y, 'ko')
plt.plot(x, reg3.predict(x3), 'ro')
plt.xlabel('number of units (in thousands)'); plt.ylabel('cost (in thousands $)'); plt.show()
```

```python
reg3.intercept_, reg3.coef_
```
